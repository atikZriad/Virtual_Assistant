```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import numpy as np
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])


df_new = df.copy()
df_new = df_new.drop_duplicates(subset=["job_title", "company_name"])

top_countires = df_new["job_country"].value_counts().head(5).index
df_job_country_salary = df_new.pivot_table(values="salary_year_avg", index="job_country", columns="job_title_short", aggfunc="median")
df_job_country_salary = df_job_country_salary.loc[top_countires]
job_title = ["Data Analyst", "Data Engineer", "Data Scientist"]
df_job_country_salary = df_job_country_salary[job_title]
df_job_country_salary.plot(kind="bar")
plt.ylabel('Median Salary ($USD)')
plt.xlabel('')
plt.title('Median Salary by Country and Job Title')
plt.xticks(rotation=45, ha='right')
plt.show()

```

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
new_df = pd.read_csv("https://lukeb.co/software_csv")
new_df.set_index("job_posted_month", inplace=True)

# Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

df_usa = df[df["job_country"] == "United States"].copy()
df_usa.insert(8, "job_posted_month", df_usa["job_posted_date"].dt.strftime("%B"))
df_usa_pivot = df_usa.pivot_table(index="job_posted_month", columns="job_title_short", aggfunc="size")
df_usa_pivot.reset_index(inplace=True)
df_usa_pivot.sort_values("job_posted_month", inplace=True)
df_usa_pivot.set_index("job_posted_month", inplace=True)

top_3 = df_usa["job_title_short"].value_counts().head(3)
top_3 = top_3.index.tolist()

# Merging
merged_top_5 = df_usa_pivot[top_3].merge(new_df, on="job_posted_month")
top_5 = (
    
    merged_top_5.sum()
    .sort_values(ascending=False)
    .head(5)
    .index
    .tolist()
)


merged_top_5[top_5].plot(kind="line")
plt.title('Monthly Job Postings for Top Tech Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.ylim(0, 20000)
plt.legend()
plt.show()



```
